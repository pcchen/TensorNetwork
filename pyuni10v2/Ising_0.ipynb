## Tensor network for partition function
* For symmetric W, in SVD one has U = V.
* Rank-4 delta tensor.

```python
# matrixes

def mat_W0(beta):
    """Create W matrix (Boltzman weight)
    """
    W = np.array([[np.exp(+beta), np.exp(-beta)],
                  [np.exp(-beta), np.exp(+beta)]])
    return W

def mat_W(beta, h):
    """Create W matrix (Boltzman weight)
    """
    W = np.array([[np.exp(+beta+beta*h), np.exp(-beta)],
                  [np.exp(-beta), np.exp(+beta-beta*h)]])
    return W

def mat_M0(beta):
    M = np.array([[np.sqrt(np.cosh(beta)), np.sqrt(np.sinh(beta))],
                  [np.sqrt(np.cosh(beta)), -np.sqrt(np.sinh(beta))]])
    return M

def mat_M(beta, h):
    W = mat_W(beta, h)
    U, S, Vd = np.linalg.svd(W)
    if U[0, 0] < 0 :
        U = -U
        Vd = -Vd
    M = U @ np.diag( np.sqrt(S) )
    return M

def mat_T(beta, h):
    T = np.zeros((2,2,2,2))
    M = mat_M(beta, h)
    for i in range(1+1):
        for j in range(1+1):
            for k in range(1+1):
                for l in range(1+1):
                    T[i,j,k,l] = M[0,i]*M[0,j]*M[0,k]*M[0,l]+M[1,i]*M[1,j]*M[1,k]*M[1,l]
    T = T.reshape((4,4))
    return T
print(mat_T(1.0, 0.0))

def delta_M():
    delta = np.zeros((4, 4))
    delta[0, 0] = 1.0
    delta[3, 3] = 1.0
    return delta

def Sz_M():
    Sz = np.zeros((4, 4))
    Sz[0, 0] = 1.0
    Sz[3, 3] = -1.0
    return Sz
# print(Sz_M())

def Szp_M():
    Sz = np.zeros((4, 4))
    Sz[0, 0] = 1.0
#     Sz[3, 3] = -1.0
    return Sz
# print(Szp_M())

def Szm_M():
    Sz = np.zeros((4, 4))
#     Sz[0, 0] = 1.0
    Sz[3, 3] = -1.0
    return Sz
# print(Szm_M())
```

```python
# bonds
bdin = uni10.Bond(uni10.BD_IN,2)
bdout= uni10.Bond(uni10.BD_OUT,2)

# tensors

# delta
delta = uni10.UniTensorR([bdin,bdin,bdout,bdout])
delta.SetName('delta')
delta.PutBlock(delta_M())
# print(delta)

# Sz
Sz = uni10.UniTensorR([bdin,bdin,bdout,bdout])
Sz.SetName('Sz')
Sz.PutBlock(Sz_M())
# print(Sz)

# Szp
Szp = uni10.UniTensorR([bdin,bdin,bdout,bdout])
Szp.SetName('Szp')
Szp.PutBlock(Szp_M())
print(Szp)

# Szm
Szm = uni10.UniTensorR([bdin,bdin,bdout,bdout])
Szm.SetName('Szm')
Szm.PutBlock(Szm_M())
print(Szm)

# M & MT
M = uni10.UniTensorR([bdin,bdout])
M.SetName('M')
M.PutBlock(mat_M(1.0, 0.0))
# print(M)

M.SetName('MT')
MT = uni10.Transpose(M)
# print(MT)

# T 
T = uni10.UniTensorR([bdin,bdin,bdout,bdout])
net = uni10.Network("W2T.net")
net.PutTensor("delta", delta)
net.PutTensor("M",M)
net.PutTensor("MT",MT)
net.Launch(T)
T.SetName('T')
print(T)

# Tz
Tz = uni10.UniTensorR([bdin,bdin,bdout,bdout])
net = uni10.Network("W2T.net")
net.PutTensor("delta", Sz)
net.PutTensor("M",M)
net.PutTensor("MT",MT)
net.Launch(Tz)
Tz.SetName('Tz')
print(Tz)

# Tzp
Tzp = uni10.UniTensorR([bdin,bdin,bdout,bdout])
net = uni10.Network("W2T.net")
net.PutTensor("delta", Szp)
net.PutTensor("M",M)
net.PutTensor("MT",MT)
net.Launch(Tzp)
Tzp.SetName('Tzp')
print(Tzp)

# Tzm
Tzm = uni10.UniTensorR([bdin,bdin,bdout,bdout])
net = uni10.Network("W2T.net")
net.PutTensor("delta", Szm)
net.PutTensor("M",M)
net.PutTensor("MT",MT)
net.Launch(Tzm)
Tzm.SetName('Tzm')
print(Tzm)
```

```python
net = uni10.Network("trace.net")
Z = uni10.UniTensorR()
net.PutTensor('T11',T)
net.PutTensor('T12',T)
net.PutTensor('T21',T)
net.PutTensor('T22',T)
net.Launch(Z)
Z = Z.GetBlock(0)[0,0]
print(Z)

net = uni10.Network("trace.net")
SzSz = uni10.UniTensorR()
net.PutTensor('T11',Tz)
net.PutTensor('T12',Tz)
net.PutTensor('T21',T)
net.PutTensor('T22',T)
net.Launch(SzSz)
SzSz = SzSz.GetBlock(0)[0,0]
print(SzSz)

net = uni10.Network("trace.net")
Szp = uni10.UniTensorR()
net.PutTensor('T11',T)
net.PutTensor('T12',Tzm)
net.PutTensor('T21',T)
net.PutTensor('T22',T)
net.Launch(Szp)
Szp = Szp.GetBlock(0)[0,0]
print(Szp)
```

```python
def Z_TN(beta):
    # M & MT
    M = uni10.UniTensorR([bdin,bdout])
    M.SetName('M')
    M.PutBlock(mat_M0(1.0))
#     print(M)

    M.SetName('MT')
    MT = uni10.Transpose(M)
#     print(MT)

    # T 
    T = uni10.UniTensorR([bdin,bdin,bdout,bdout])
    T.SetName('T')

    net = uni10.Network("W2T.net")
    net.PutTensor("delta", delta)
    net.PutTensor("M",M)
    net.PutTensor("MT",MT)
    net.Launch(T)
#     print(T)
    
    net = uni10.Network("trace.net")
    Z = uni10.UniTensorR()
    net.PutTensor('T11',T)
    net.PutTensor('T12',T)
    net.PutTensor('T21',T)
    net.PutTensor('T22',T)
    net.Launch(Z)
    
    Z = Z.GetBlock(0)[0,0]
#     print(Z)
    return Z
```

```python
def contract(beta, h):
    # M & MT
    M = uni10.UniTensorR([bdin,bdout])
    M.SetName('M')
    M.PutBlock(mat_M(beta, h))
#     print(M)
    M.SetName('MT')
    MT = uni10.Transpose(M)
#     print(MT)

    # T 
    T = uni10.UniTensorR([bdin,bdin,bdout,bdout])
    T.SetName('T')
    net = uni10.Network("W2T.net")
    net.PutTensor("delta", delta)
    net.PutTensor("M",M)
    net.PutTensor("MT",MT)
    net.Launch(T)
#     print(T)
    
    net = uni10.Network("trace.net")
    Z = uni10.UniTensorR()
    net.PutTensor('T11',T)
    net.PutTensor('T12',T)
    net.PutTensor('T21',T)
    net.PutTensor('T22',T)
    net.Launch(Z)
    Z = Z.GetBlock(0)[0,0]

    net = uni10.Network("trace.net")
    SzSz = uni10.UniTensorR()
    net.PutTensor('T11',Tz)
    net.PutTensor('T12',Tz)
    net.PutTensor('T21',T)
    net.PutTensor('T22',T)
    net.Launch(SzSz)
    SzSz = SzSz.GetBlock(0)[0,0]
#     print(SzSz)
    
    return Z, SzSz

print(contract(1, 0))
```
